```python
%reset -f

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('logExoANDNet_c213b90a14211698.csv')
df
```

```python
array_full = np.array(df)
array_full.shape[0]
```

```python
time_line = [i for i in range(array_full.shape[0])]
```

```python
# seconds_only = array_full[0,0][17:19]
```

```python
plt.figure(figsize=(8, 6), dpi=120)
plt.grid(True, color='yellow', linestyle='-', linewidth=0.4)
plt.gca().set_facecolor('black')
plt.title('Graf mjerenja')

# Parameters
plt.plot(time_line, array_full[:,1], label='RSSI', linewidth=2)
plt.plot(time_line, array_full[:,2], label='RSRP', linewidth=2)
plt.plot(time_line, array_full[:,3], label='RSRQ', linewidth=2)
plt.plot(time_line, array_full[:,4], label='RSSNR', linewidth=2)

# Axis labels
plt.xlabel('Time')
plt.ylabel('Parameters')

plt.legend(loc=0)
plt.show()
```

```python
plt.figure(figsize=(8, 6), dpi=120)
plt.grid(True, color='yellow', linestyle='-', linewidth=0.4)
plt.gca().set_facecolor('black')
plt.title('Graf mjerenja')

# Parameters
plt.plot(time_line, array_full[:,9], label='Bitrate (DataSource)', linewidth=2)
# plt.plot(time_line, array_full[:,11], label='Resolution (Displayed)', linewidth=2)
plt.plot(time_line, array_full[:,13], label='Rebuffering', linewidth=2)

# Axis labels
plt.xlabel('Time')
plt.ylabel('Parameters')

plt.legend(loc=0)
plt.show()
```
